```python
import code_github_range_integers as func
```

## Initial Variables

```python
nqubits = 4
qubits_register = QuantumRegister(nqubits, name='q')
```

```python
approx_QFT = 0
```

```python
low_boundary = 4
upper_boundary = 7
```

## Two circuits

```python
circuitA = QuantumCircuit(qubits_register)
circuitB = QuantumCircuit(qubits_register)
```

## Implementation A

```python
circuitA.h(qubits_register)
```

```python
circuitA.append(func.oracle_interval_A(target_reg=qubits_register, low_boundary=low_boundary, upper_boundary=upper_boundary),
              qubits_register)
```

```python
circuitA.decompose(reps=0).draw(output='mpl')
```

## Circuit B

## Results using Statevector

```python
psiA = qi.Statevector.from_instruction(circuitA)
```

```python
for n, v in enumerate(np.asarray(psiA)):
    print('%d: %.2f' %(n, v.real))
```

```python
psiB = qi.Statevector.from_instruction(circuitB)
```

```python
for n, v in enumerate(np.asarray(psiB)):
    print('%d: %.2f' %(n, v.real))
```
